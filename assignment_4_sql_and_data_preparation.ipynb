## Cleaning dataset

One way to deal with missing data is to replace it with mean value of that column.
This works especially with the missing age values as it doesnt skew the data.

```python

titanic['age'] = titanic['age'].fillna(titanic['age'].median())

titanic['age'].notna()
```
