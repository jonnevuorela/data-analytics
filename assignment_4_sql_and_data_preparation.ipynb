## Cleaning dataset

One way to deal with missing data is to replace it with mean value of that column.
Many sources state that this is a good way to deal with a missing age value.
As one redditor states. Data tends to be similiar to the mean.

```python


print("mean pre fill")
print(titanic['Age'].mean())

titanic['Age'] = titanic['Age'].fillna(titanic['Age'].mean())


print("mean after fill")
titanic['Age'].mean()
```


Dealing with the missing age values by this way seems also not to affect the mean value.


```python

titanic['Cabin'] = titanic['Cabin'].fillna('Unknown')
```


I think its feasible to fill the blank cabin values with unknown since its not going to do much difference.
I found a notebook on kaggle which stated that survival rates had correlation with decks,
since higher class passengers had cabins on decks A-C and lower class had cabins on decks E-G.

Ticket prefixes correlate weakly with survival. ticket frequency is more useful.

```python

# Regex pattern to match abbreviations
titanic['Title'] = titanic['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False)
print(titanic['Title'].value_counts())

print(titanic.groupby('Title')['Survived'].mean().sort_values(ascending=False))
```
