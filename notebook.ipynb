## Section 3-4

```python
import statistics
import math

data_types = {
    "discrete": [0, 1, 2, 3],                      
    "continuous": [1.23, 3.5, 7.8],                
    "nominal": ["red", "green", "blue"],          
    "ordinal": ["low", "medium", "high"]           
}


discrete_mean = statistics.mean(data_types["discrete"])
continuous_mean = statistics.mean(data_types["continuous"])
print("Discrete mean:", discrete_mean)
print("Continuous mean:", continuous_mean)

ordinal_mode = statistics.mode(data_types["ordinal"])
print("Ordinal mode:", ordinal_mode)

print("math.sqrt(discrete_mean):", math.sqrt(discrete_mean))
print("math.ceil(continuous_mean):", math.ceil(continuous_mean))
print("math.fsum(continuous):", math.fsum(data_types["continuous"]))

print("math.prod(discrete):", math.prod(data_types["discrete"]))
```

### Virtual environments
There is an ever long problem with scripting languages and portability, and python doesn't make it any easier being part of all the other relevant PC systems apart from Windows. <br>
Virtual enviroments is great way to tackle pythons poor portability, by offering a way to maintain a contained distribution of dependecies and interpreter version. Paired with runtime version managing tool like mise, we can achieve well organized development or research environment regardless of the system. 

### Type hints
Generally speaking when I talk about my personal preferences, I __love__ strongly typed programming languages.<br>
In my opinion there is no situation when we should even be able to be unsure about some values type while trying to apply logic in to the program. Therefore I think that the compile time is a perfect place to catch these humane errors. <br><br>
Then we have these interpreted technologies with often no enforcing on types whatsoever, let alone the tooling for type checking even be provided. I get that it is not in the nature of scripting to be concerned about types or possible failures, but I see it in the way that I would like to know when I personally make a mistake while programming and not get the program to be like 'i dunno' or 'nah' when asked for an answer.<br>
But it's not just the scripting and prototyping that is riddled with this problem but also the ever growing usage of technologies like python or god forbid javascript on building backends and native software. All in all it's a developers responsibility to pick a suitable technology for the job.<br><br>
That being said, it's a good thing in my opinion that there is typing tools available even if its not enforced to use.
